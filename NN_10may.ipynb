```
#### Assigning Weights to each inputs
np.random.seed(10)
weights = np.random.random((3,1))
weights
```

```
def derivative(x):
  return x *(1-x) ##derivative for sigmoid
```

```
for i in range(300):
  input_layer =input_samples
  mul=np.dot(input_layer,weights) ##dot multiplication 
  outputs = activation(mul)
  error=output_samples-outputs
  adjustments =error * derivative(outputs) #gradient desent
  weights +=np.dot(input_layer.T,adjustments) # back Propagation 
  if i%5 == 0 :
    print(error)
    print("\n")
```

```
### Create NN for XOR Gate Using Tensorflow 
import numpy as np

import tensorflow as tf
```

```

input_samples = np.array([[0,0],[0,1],[1,0],[1,1]])
output_samples = np.array([[0],[1],[1],[0]])
```

```
input_samples 
```

```
output_samples
```

```
model = tf.keras.models.Sequential()
```

```
## adding layers 2 
model.add(tf.keras.layers.Dense(len(input_samples),input_dim=2,activation='relu')) # input layer 
```

```
## adding the layers (2 layers)
model.add(tf.keras.layers.Dense(1,activation= 'sigmoid')) #output layer
```

```
# compile the model
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])
```

```
#### Train the model
trained_model = model.fit(input_samples,output_samples,epochs=900)
```

```
import matplotlib.pyplot as plt
loss_curve = trained_model.history['loss']
plt.plot(loss_curve)
plt.show()
```

```
import matplotlib.pyplot as plt
accu_curve = trained_model.history['binary_accuracy']
plt.plot(accu_curve)
plt.show()
```

```
model.predict(input_samples).round()
```

```
model.save('mymodel.model')
```
